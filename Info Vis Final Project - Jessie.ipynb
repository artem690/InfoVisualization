## Which airlines are the safest to fly on?

```python
alt.Chart(airData).mark_circle().encode(
    x = "incidents_85_99", 
    y = "incidents_00_14",
    size = "incidents_00_14",
    tooltip = ["airline", "incidents_85_99", "incidents_00_14"]
).interactive()
```

The above visualization suggests that travelers may want to avoid falying on Delta/Northwest, American, United/Continental, and US Airways/America West since they have the highest number of incidents in the more current time period.

```python
alt.Chart(airData).mark_circle().encode(
    x = "avail_seat_km_per_week", 
    y = "incidents_00_14",
    size = "avail_seat_km_per_week",
    tooltip = ["airline", "incidents_00_14", "avail_seat_km_per_week"],
    color = alt.Color("incidents_00_14", scale = alt.Scale(scheme="magma"))
)
```

This visualization shows the number of incidents that occured between 2000 and 2014 with the number of available seat kilometers traveled for each airline. This is a more accurate representation because some airlines fly more than others, which means it makes sense for those airlines to have more incidents. By looking at this graph, we see that Delta/Northwest, United/Continental, and American had the most incidents, but their available seat kilometers were much higher than other airlines. This visualization suggests that Pakistan International and Saudi Arabian airlines may not be the safest to fly on since the ratio of incidents to available seat kilometers is relatively high.

Here we are just finding out what is the percentage of accidents are fatal accidents, that will help us to look at the data closer and understand it more. We are looking at 2 different periods of the time that we have in our dataset.

```python
# dict = {"Airline": [], "Fatalities %": []}
dictt = {} 
for i, j in airData.iterrows():
    if j['incidents_85_99'] == 0:
        continue
    else:
        perc = round(100 * (j['fatal_accidents_85_99'] / j['incidents_85_99']))
        dictt[j['airline']] = perc

# num = []
# airlines = []
for i in dictt:
#     num.append(dictt[i])
#     airlines.append(i)
    print(i, "has", dictt[i], "% of fatal incidents in between 1985 and 1995")

    
```

```python
# dict = {"Airline": [], "Fatalities %": []}
dictt = {} 
for i, j in airData.iterrows():
    if j['incidents_00_14'] == 0:
        continue
    else:
        perc = round(100 * (j['fatal_accidents_00_14'] / j['incidents_00_14']))
        dictt[j['airline']] = perc

for i in dictt:
    print(i, "has", dictt[i], "% of fatal incidents in between 2000 and 2014")
    
```
